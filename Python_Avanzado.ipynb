```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(
    r'datasets/Info_pais.csv',
    encoding='ISO-8859-1',
    delimiter=';',
    decimal=',' # Decimal es la opción para que lo entienda como número poniendo un punto en lugar de coma. Sino lo interpreta como strings.
)
```

```python
df
```

```python
df_order = df.sort_values(
    'Poblacion',
    ascending=False
 )

df_order.head()
```

```python
 plt.scatter(df_order['Renta per capita'], df_order['Esperanza de vida']);
# Con Jupyter se pueden copiar y pegar o arrastrar los gráficos como imágenes.
# Hay que añadir título, y etiquetas de los ejes (variables del gráfico)
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida']
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida');
```

```python
# IMPORTANTÍSIMO!

# Se crea una columna nueva normalizando la población, para cuando la represente no haya tapados entre unos y otros por diferencias muy notables de tamaño.
# Así que se normaliza para representarlo mejor.

df_order['Poblacion_normalizada'] = df_order['Poblacion']/max(df_order['Poblacion']) # La población de cada país entre la población de china, que es la que tienen máxima población.
```

```python
df_order
```

```python
# Sigue habiendo mucha diferencia para representarlo. Puede mejorarse un poquito más.

df_order['Poblacion_normalizada'] = df_order['Poblacion']/(max(df_order['Poblacion'])/10000) 
# Lo mismo pero ahora la población de china (la máxima) entre 10000.
# Se hace amenudo entre qué número dividir la máxima para normalizar, si 1000, 10000 o 100000. Porque tal tamaño en 1, como estaba antes, es una bola muy pequeña siendo esta la más grande.

df_order
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s = df_order['Poblacion_normalizada'] # s es de size, de tamaño.
 )

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per capita')
plt.ylabel('Esperanza de vida');
```

```python
# Aparece un problema de inundación, que unas bolas tapan otras, pero el problema real es que son del mismo color.
# Las bolas están muy pegadas. Habría que hacer más grande el gráfico, pero no las bolas.
# Por lo que hay que agrandar las dimensiones del gráfico.
```

```python
plt.scatter(
     df_order['Renta per capita'],
     df_order['Esperanza de vida'],
     s = df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf() # Grafic configuration
fig.set_size_inches(14.5, 10); # Configura el tamaño de las pulgadas. (Alto, Ancho)
```

```python
plt.scatter(
     df_order['Renta per capita'],
     df_order['Esperanza de vida'],
     s = df_order['Poblacion_normalizada'],
     c = df_order['Poblacion_normalizada'] # c de color.
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10);
```

```python
plt.scatter(
     df_order['Renta per capita'],
     df_order['Esperanza de vida'],
     s = df_order['Poblacion_normalizada'],
     c = df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10);

df_order.reset_index(drop=True, inplace=True) # Drop = True significa que el índice que había antes lo tire, que lo ponga ordenado.
# E implace True es que guarde esos cambios.
for i in range(0, 10):
     plt.annotate( # Anotar en el gráfico, porque es plt.
         df_order['País'][i],
         (
             df_order['Renta per capita'][i],
             df_order['Esperanza de vida'][i]
         )
     )
# Todo esto es para escribir el nombre respectivo de los 10 primero países ordenador por población.
```

```python
plt.scatter(
     df_order['Renta per capita'],
     df_order['Esperanza de vida'],
     s = df_order['Poblacion_normalizada'],
     c = df_order['Poblacion_normalizada'],
     alpha = 0.5 # Es la transparencia.
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

plt.colorbar() # Es un índice de color por población, en lugar de tener un color por país que es lioso, este puede ser una buena forma de identificar.
fig = plt.gcf()
fig.set_size_inches(14.5, 10);

df_order.reset_index(drop=True, inplace=True)

for i in range(0, 10):
    plt.annotate( # si quieres escribir solo uno, necesita que les des un dato para eje X y otro para eje Y. Qué quierees pintar y donde (X, Y).
        df_order['País'][i],
        (
             df_order['Renta per capita'][i],
             df_order['Esperanza de vida'][i]
         )
     )
```

```python
import random
import pandas as pd
```

```python
mascotas = []
dataset = []
df = None
```

```python
def set_codigos_mascotas(numero_de_mascotas):
    
    mascota_prefix = ["P"]*5 + ["G"]*4 + ["H"]*1  # 50% Perro / 40% Gato / 10% Hurón.
    sexo = ["M"]*5 + ["H"]*5  # 50% Macho / 50% Hembra
    colores = ["BLA"]*35 + ["MAR"]*25 + ["GRI"]*14 + ["NEG"]*26  # 35% BLANCO / 25% MARRÓN / 14% GRIS / 26% NEGRO
    grupo_alpha = []  # Solo se añade la parte alfabética del código a generar.
    
    for i in range(numero_de_mascotas):
        color = random.choice(colores)
        mascota = random.choice(mascota_prefix) + random.choice(sexo)
        mascota += color
        grupo_alpha.append(mascota)
        mascota += str(grupo_alpha.count(mascota)).zfill(2)  # zfill añade ceros por la izquierda hasta completar la cadenza de caracteres que le asignemos en los paréntesis.
        mascotas.append(mascota)
```

```python
def get_mascota(codigo_mascota):
    
    if codigo_mascota.startswith("P"):
        return "Perro"
    elif codigo_mascota.startswith("G"):
        return "Gato"
    elif codigo_mascota.startswith("H"):
        return "Hurón"
    else:
        return "Desconocido"
```

```python
def get_sexo(codigo_mascota):
    
    if codigo_mascota[1] == "M":
        return "Macho"
    elif codigo_mascota[1] == "H":
        return "Hembra"
    else:
        return "Desconocido"
```

```python
def get_color(codigo_mascota):
    
    if codigo_mascota[2:5] == "BLA":
        return "Blanco"
    elif codigo_mascota[2:5] == "MAR":
        return "Marrón"
    elif codigo_mascota[2:5] == "GRI":
        return "Gris"
    elif codigo_mascota[2:5] == "NEG":
        return "Negro"
    else:
        return "Desconocido"
```

```python
def set_dataframe(numero):
    
    global df  # Porque la hemos creado fuera sin ningún valor y en esta función en concreto le vamos a crear un valor.
    # De esta manera decimos: Esta variable de fuera es global si se va a utilizar aquí.
    set_codigos_mascotas(numero)
    
    for codigo_mascota in mascotas:
        dataset.append({
            "Codigo mascota": codigo_mascota,
            "Mascota": get_mascota(codigo_mascota),
            "Sexo": get_sexo(codigo_mascota),
            "Color": get_color(codigo_mascota),
            "Numero": int(codigo_mascota[-2:])
        })
    
    df = pd.DataFrame(dataset)
```

```python
set_dataframe(100)
```
